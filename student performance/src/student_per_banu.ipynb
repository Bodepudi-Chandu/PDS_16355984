```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tips = sns.load_dataset("tips")

# Create a boxplot showing the distribution of total bill amounts for each day
sns.boxplot(x="day", y="total_bill", data=tips)

# Set the title and labels
plt.title('Distribution of Total Bill Amounts by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount')

plt.savefig('/content/snaps_of_results/Total_Bill_Distribution_Boxplot.jpg')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tips = sns.load_dataset("tips")

# Create a scatter plot showing the relationship between total bill amount and tip amount
sns.scatterplot(x="total_bill", y="tip", data=tips)

# Set the title and labels
plt.title('Total Bill Amount vs. Tip Amount')
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')

plt.savefig('/content/snaps_of_results/Total_Bill_vs_Tip_Scatterplot.jpg')
plt.show()
```

```python

# visualization 4: Density plot of Reading Scores by Test preparation course
sns.kdeplot(x='reading score', data=data, hue='parental level of education', fill=True, alpha=.3)
plt.title('Reading Scores by Test prep')
plt.xlabel('Reading Score')
plt.savefig ('/content/snaps_of_results/Reading_score_by_ParentalLevelEducation_course.png')
plt.show()
```

```python
plt.hist([data['writing score'],data['math score'],data['reading score']], bins=20,  color=['red','green','blue'], edgecolor='black')
plt.xlabel('All 3 types of score')
plt.ylabel('Values')
plt.title('Histogram of All 3 types of score')
plt.show()
plt.savefig('/content/snaps_of_results/histograh.PNG')
```

```python
import matplotlib.pyplot as plt

# Count the number of occurrences of each gender
gender_counts = data['lunch'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts, color=['#1f77b4', '#ff7f0e'])
plt.title('Type of Lunch Percentage')
plt.xlabel('Lunch Type')
plt.ylabel('Count')

# Adding percentage labels on top of each bar
total = sum(gender_counts)
for i, count in enumerate(gender_counts):
    plt.text(i, count + 5, f'{count/total*100:.1f}%', ha='center')

plt.savefig('/content/snaps_of_results/Lunch_type_percentage_barplot.jpg')
plt.show()
```
